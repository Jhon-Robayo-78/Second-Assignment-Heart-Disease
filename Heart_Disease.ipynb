# Construcción del modelo

### Estimación del modelo de regresión logística. 

```python
X = data[['painexer','cp','proto','slope','ladprox','laddist','diag','cxmain','ramus','om1','om2','rcaprox','rcadist']] # variables predictoras
Y = data[['num']] # variable objetivo
```

### Asignación de datos para entrenamiento y testeo 

```python
# division de los datos de entrenamiento(70%) y test(30%)
X_train, X_test, Y_train , Y_test =train_test_split(X,Y,test_size=0.3, random_state=0)
```

##### Instanciamiento de la clase de regresión logíca

```python
log_regression = LogisticRegression()
```

#### Asignación de datos de entrenamiento 

```python
log_regression.fit(X_train,Y_train)
```

#### modelo de predicción de datos

```python
y_pred = log_regression.predict(X_test)
y_pred
```

valores de confusión del modelo

```python
cnf_matrix=metrics.confusion_matrix(Y_test,y_pred)
cnf_matrix
```

### Medida de presición del modelo expuesto:

```python
print("Accuracy: ",metrics.accuracy_score(Y_test,y_pred))
```
